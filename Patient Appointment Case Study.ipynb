```python
#renaming columns
df.rename(columns={'Scholarship':'Sponsorship'},inplace=True)
df.rename(columns={'Hipertension':'Hypertension'},inplace=True)
df.rename(columns={'Handcap':'Handicap'},inplace=True)
```

```python
df
```

```python
#correction of data types
df['ScheduledDay']=df['ScheduledDay'].astype('datetime64[ns]')
df['AppointmentDay']=df['AppointmentDay'].astype('datetime64[ns]')
df['Sponsorship']=df['Sponsorship'].astype('bool')
df
```

```python
#Renaming No-show column to Showed-up and making sure records corresponds to column name
df['Showed-up']=df['No-show'].apply(lambda x: 'True' if x=='No' else 'False')
```

```python
def myplot(a,b):
    df.groupby([a])['PatientId'].count().plot(kind=b)
    plt.title(('Patients Per '+ a))
    plt.xlabel(a)
    plt.ylabel('Total Number of Patients')
    c=df.groupby([a])['PatientId'].count()
    print (c)
    
```

```python
#Patients with Hypertension Per their Show up status
Hs=df.groupby('Hypertension')['Showed-up'].value_counts().unstack('Showed-up')
Hst=Hs.loc[True].rename('Hypertension').to_frame()
```

```python
Hst
```

```python
#Patients with Diabetes Per their Show up status
Ds=df.groupby('Diabetes')['Showed-up'].value_counts().unstack('Showed-up')
Dst=Ds.loc[True].rename('Diabetes').to_frame()
Dst
```

```python
J=Hst.join(Dst)
J
```

```python
J.plot(kind='bar')
```

```python
#Patients with Alcoholism Per their Show up status
As=df.groupby('Alcoholism')['Showed-up'].value_counts().unstack('Showed-up')
Ast=As.loc[True].rename('Alcoholism').to_frame()
Ast
```

```python
H=J.join(Ast)
H.plot(kind='bar',figsize=[12,6])
```
